```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list_MSFT=[]
```

```python
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#MSFT since:2021-01-01 until:2021-09-21 lang:en').get_items()):
    tweets_list_MSFT.append([tweet.date, tweet.content])
```

```python
jan_sept = pd.DataFrame(tweets_list_MSFT, columns=['Datetime', 'Text'])
jan_sept.to_csv(r'C:\Users\sowmy\Desktop\jan_sept_MSFT.csv', index=False)
```
